## 1. Project Objective and Research Questions
### Project Objective
The objective of this project is to analyse how Reddit users discussed the release of *The Devil Wears Prada 2*. The project combines social network analysis and content analysis to identify discussion communities, central users, sentiment patterns, emotional reactions and dominant named entities across different Reddit audiences.

### Research Question
How did Reddit users engage with the release of *The Devil Wears Prada 2*, and how do discussion communities, sentiment, and dominant themes differ across film, fashion, celebrity-culture, and box-office audiences?

### Sub-questions
- RQ1 — Network structure: Which discussion communities emerge from the Reddit reply network?
- RQ2 — Centrality: Which users are the most central in the discussion, and which users act as bridges between communities?
- RQ3 — Sentiment and emotions: How do sentiment and emotional patterns differ across detected communities and subreddits?
- RQ4 — Named entities and narratives: Which people, brands, places, organizations and works are most frequently mentioned in each community?
